```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import joblib

# Load the dataset
df = pd.read_csv("C:\\Users\\Mayuresh_kadale.DESKTOP-UBRIRCI\\Desktop\\waterways_project\\expanded_india_agriculture_data.csv")

# Features (inputs) and target (output)
X = df[['State', 'Soil Type', 'Crop Type']]
y = df['Conservation Technique']

# Encode categorical variables
label_encoders = {}
for column in X.columns:
    le = LabelEncoder()
    X.loc[:, column] = le.fit_transform(X[column])  # Use loc to avoid SettingWithCopyWarning
    label_encoders[column] = le

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Function to make predictions based on input
def predict_conservation_technique(state, soil_type, crop_type):
    # Check if the inputs are valid
    if state not in label_encoders['State'].classes_:
        return "Unknown state. Please provide a valid state."
    if soil_type not in label_encoders['Soil Type'].classes_:
        return "Unknown soil type. Please provide a valid soil type."
    if crop_type not in label_encoders['Crop Type'].classes_:
        return "Unknown crop type. Please provide a valid crop type."
    
    # Convert inputs to numerical form using label encoders
    state_encoded = label_encoders['State'].transform([state])[0]
    soil_type_encoded = label_encoders['Soil Type'].transform([soil_type])[0]
    crop_type_encoded = label_encoders['Crop Type'].transform([crop_type])[0]
    
    # Create a DataFrame for the prediction input with the same columns as the training data
    input_df = pd.DataFrame([[state_encoded, soil_type_encoded, crop_type_encoded]], 
                            columns=['State', 'Soil Type', 'Crop Type'])
    
    # Predict conservation technique
    prediction = clf.predict(input_df)  # Pass DataFrame with feature names
    return prediction[0]

# Example usage
state = 'Andhra Pradesh'
soil_type = 'Loamy'
crop_type = 'Rice'
recommended_technique = predict_conservation_technique(state, soil_type, crop_type)
print(f"Recommended Water Conservation Technique: {recommended_technique}")

# Save the trained model
joblib.dump(clf, 'model.pkl')

# Save the label encoders
for col, le in label_encoders.items():
    joblib.dump(le, f'{col}_encoder.pkl')
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Feature importance from the decision tree model
feature_importance = clf.feature_importances_

# Plotting the feature importance
plt.figure(figsize=(8,6))
plt.barh(X.columns, feature_importance, color='skyblue')
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance in Decision Tree')
plt.show()
```

```python
from sklearn.model_selection import learning_curve
import numpy as np

# Learning curve
train_sizes, train_scores, test_scores = learning_curve(clf, X, y, cv=5, scoring='accuracy', n_jobs=-1)

# Calculate mean and standard deviation
train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

# Plotting the learning curve
plt.figure(figsize=(8,6))
plt.plot(train_sizes, train_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, test_mean, 'o-', color="g", label="Cross-validation score")
plt.xlabel('Training Examples')
plt.ylabel('Score')
plt.title('Learning Curve')
plt.legend(loc="best")
plt.show()
```
